**Now add *multiple* predator and prey agents ($N > 1$ for both). Find a combination of $N_{\mathrm{Prey}}$ and $N_{\mathrm{Predator}}$ where the Prey can maintain a sustainable population (I.e., they aren't wiped out in the long run).**

```python
import random

def running_environment_with_predator(n_prey, n_predator, tot_time=40):
    my_environment = Environment()
    for _ in range(n_prey):
        #chosing male or female 50/50
        sex = random.choice(["M", "F"])
        #making prey
        prey_agent = Animal_Agent(ptype="Prey", sex=sex)
        #choosing alleles
        allele_1 = random.choice([0,1])
        allele_2 = random.choice([0,1]) 
        #setting alleles
        prey_agent.set_allele(allele_1, allele_2)
        my_environment.add_agent(prey_agent)

    #creating predators 
    for _ in range(n_predator):
        predator_agent = Animal_Agent(ptype="Predator", shape="P", color="red")
        my_environment.add_agent(predator_agent)

    my_environment.simulate(tot_time)
    my_environment.simulate_plot_populations(tot_time)
    

running_environment_with_predator(n_prey=80, n_predator=10)
```

## 2. Refining the Model

For this part, we're going to use our generic `Animal_Agent`class to construct child classes that represent *specific* animal species for both our predator and prey animals. Specifically, we'll be making a `Mouse` child class for the prey and a `Hawk` child class for the predator. **We'll use these new child classes to run our natural selection model.**

### 2.1 (15 Points)

Create a new `Mouse` child class, using the `Animal_Agent` as the parent class. The `Mouse` class will largely be the same as the `Animal_Agent` class except for two details: 

1. **In the `set_allele` method, add a new class attribute called `catch_probability`. For the light-brown mice, set `catch_probability` to 0.95. For dark brown mice, set the `catch_probability` to 0.05.**
2. **Change the `Procreate` method so that they create new `Mouse` animal objects.**



```python
import numpy as np

#Mouse child class
class Mouse(Animal_Agent):
    #constructor
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.ptype = "Prey"
        self.catch_probability = 0.05

    #altered set allele
    def set_allele(self, alle_d, alle_m):
        super().set_allele(alle_d, alle_m)
        if self.alle_d + self.alle_m >= 1:
            # Probability of getting caught for light brown mice
            self.catch_probability = 0.95
        else:
            # probability of getting caught for dark brown mice
            self.catch_probability = 0.05
         
    #altered procreate
    def procreate(self, all_agents, time):
        child_list = []
        val = random.random()
        if (self.sex == 'F') and (np.mod(abs(time - self.last_litter_time), self.gestation) == 0):
            # if any male mice are close enough, mate
            for agent in all_agents:
                if (agent.ptype == "Prey") and (agent.sex == "M") and (self.get_distance(agent) <= self.mate_range) and (val > (1.0 / (1.0 + np.exp(-len(all_agents) / self.saturation_pop)))):
                    child_num = int(np.random.normal(7, 1))
                    for jj in range(child_num):
                        mom_a = self.alle_d
                        if np.random.randint(0, 2) == 1:
                            mom_a = self.alle_m
                        dad_a = agent.alle_d
                        if np.random.randint(0, 2) == 1:
                            dad_a = agent.alle_m
                        sex = "F"
                        if np.random.randint(0, 2) == 1:
                            sex = "M"
                        #this part makes a mouse instead of animal_agent
                        child = Mouse(sex=sex)
                        child.set_allele(dad_a, mom_a)
                        child_list.append(child)
                    self.last_litter_time = time
                    break

        return child_list
```

### 2.2 (10 Points)

Create a new `Hawk` child class, using the `Animal_Agent` as the parent class. The `Hawk` class will largely be the same as the `Animal_Agent` class except for one detail: **In the `hunt` method, incorporate the `catch_probability` of the predator, such that it is harder for a `Hawk` to catch a `Mouse` with a lower `catch_probability`.**

```python
#The Hawk child class
class Hawk(Animal_Agent):
    #constructor
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        #this animal is a predator
        self.ptype = "Predator"

    #altered hunt method
    def hunt(self, all_agents):
        for agent in all_agents:
            if (agent.ptype == "Prey") and (agent.age < agent.age_limit) and not (agent.is_caught):
                if self.get_distance(agent) < self.catch_radius:
                    catch = random.random()
                    #checking if we caught them 
                    if agent.catch_probability:
                        if catch <= agent.catch_probability:
                            agent.is_caught = True
                            return agent
                        
        #return false if didn't catch any
        return False
```

### 2.3 (5 Points)

```python
def running_environment_with_Hawk_Mice(n_prey, n_predator, tot_time=40):
    my_environment = Environment()
    for _ in range(n_prey):
        sex = random.choice(["M", "F"])
        prey_agent = Mouse(sex=sex)
        allele_1 = random.choice([0,1])
        allele_2 = random.choice([0,1])        
        prey_agent.set_allele(allele_1, allele_2)
        my_environment.add_agent(prey_agent)

    for _ in range(n_predator):
        predator_agent = Hawk(shape="P", color="red")
        my_environment.add_agent(predator_agent)

    my_environment.simulate(tot_time)
    my_environment.simulate_plot_populations(tot_time)
    

running_environment_with_Hawk_Mice(n_prey=80, n_predator=10)
```
