```python
def running_environment_with_predator():
    my_environment = Environment()

    female_agent = Animal_Agent(ptype="Prey", sex="F")
    female_agent.set_allele(1, 0)

    my_environment.add_agent(female_agent)

    predator_agent = Animal_Agent(ptype="Predator", shape="P", color="red")

    my_environment.add_agent(predator_agent)

    my_environment.simulate(tot_time=20)

# Call the function
running_environment_with_predator()
```

### 1.5 (15 Points)

**Now add *multiple* predator and prey agents ($N > 1$ for both). Find a combination of $N_{\mathrm{Prey}}$ and $N_{\mathrm{Predator}}$ where the Prey can maintain a sustainable population (I.e., they aren't wiped out in the long run).**

- **NOTE 1:** Make sure you add both male and female Prey, so they can `Procreate`. 
- **NOTE 2:** We recommend using the `simulate_plot_populations()` method, so you can monitor the growth/decline of populations.

```python
import random

def running_environment_with_predator(n_prey, n_predator, tot_time=40):
    my_environment = Environment()
    for _ in range(n_prey):
        sex = random.choice(["M", "F"])
        prey_agent = Animal_Agent(ptype="Prey", sex=sex)
        allele_1 = random.choice([0,1])
        allele_2 = random.choice([0,1])        
        prey_agent.set_allele(allele_1, allele_2)
        my_environment.add_agent(prey_agent)

    for _ in range(n_predator):
        predator_agent = Animal_Agent(ptype="Predator", shape="P", color="red")
        my_environment.add_agent(predator_agent)

    my_environment.simulate(tot_time)
    my_environment.simulate_plot_populations(tot_time)
    

running_environment_with_predator(n_prey=80, n_predator=10)
```
